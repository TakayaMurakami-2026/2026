# 第3章: 正規表現
Wikipediaの記事を以下のフォーマットで書き出したファイル[jawiki-country.json.gz](/data/jawiki-country.json.gz)がある。

* 1行に1記事の情報がJSON形式で格納される
* 各行には記事名が"title"キーに、記事本文が"text"キーの辞書オブジェクトに格納され、そのオブジェクトがJSON形式で書き出される
* ファイル全体はgzipで圧縮される

以下の処理を行うプログラムを作成せよ。

## 20. JSONデータの読み込み
Wikipedia記事のJSONファイルを読み込み、「イギリス」に関する記事本文を表示せよ。問題21-29では、ここで抽出した記事本文に対して実行せよ。

```python
with open("jawiki-country.json.gz", "r") as f:
  lines=f.readlines()
  for line in lines:
    if "イギリス" in line:
      print(line)
```

```python
import json

with open("jawiki-country.json.gz", "r") as f:
    for line in f:
        data = json.loads(line)
        if data["title"] == "イギリス":
            print(data["text"])
            break
```

```python
import gzip
import json

with gzip.open("jawiki-country.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
        data = json.loads(line)
        if data["title"] == "イギリス":
            print(data["text"])
```

## 21. カテゴリ名を含む行を抽出
記事中でカテゴリ名を宣言している行を抽出せよ。

```python
import gzip
import json

with gzip.open("jawiki-country.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
      data = json.loads(line)
      if "Category" in data["text"]:
        print(data["text"])
        break
```

```python
import gzip
import json

with gzip.open("jawiki-country.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
      data = json.loads(line)

      for line in data["text"].split("\n"):
        if "Category" in line:
          print(line)
```

## 22. カテゴリ名の抽出
記事のカテゴリ名を（行単位ではなく名前で）抽出せよ。

```python
import gzip
import json

with gzip.open("jawiki-country.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
      data = json.loads(line)

      for line in data["text"].split("\n"):
        if "Category" in line:
          print(line.split(":")[1].split("|")[0].replace("]]", ""))
```

## 23. セクション構造
記事中に含まれるセクション名とそのレベル（例えば"== セクション名 =="なら1）を表示せよ。

```python
import gzip
import json
import re

with gzip.open("jawiki-country.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
      data = json.loads(line)

      for lines in data["text"].split("\n"):
        match = re.match(r"(=+)\s*(.*?)\s*\1", lines)
        if match:
          level = len(match.group(1)) - 1
          name = match.group(2)
          print(f"{name} : {level}")
```

## 24. ファイル参照の抽出
記事から参照されているメディアファイルをすべて抜き出せ。

```python
import re
import gzip
import json

with gzip.open("jawiki-country.json.gz", "rt", encoding="utf-8") as f:
    for line in f:
      data = json.loads(line)
      for lines in data["text"].split("\n"):
        matches = re.findall(r"\[\[ファイル:(.*?)(?:\|.*?)?\]\]", lines)
        for m in matches:
          print(m)
```
